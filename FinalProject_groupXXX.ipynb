```python
import os
import nibabel as nib
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
#load all flair contrast images from BRATS 2018
flair_path = glob.glob('MICCAI_BraTS_2018_Data_Training/**/**/*_flair.nii')+ glob.glob('MICCAI_BraTS_2018_Data_Validation/**/*_flair.nii')
flair_2018 = []
for img in flair_path:
    image_flair=nib.load(img).get_fdata()
    flair_2018.append(image_flair)
len(flair_2018)
```

```python
#load segmented ground truth from BRATS 2018
seg_path = glob.glob('MICCAI_BraTS_2018_Data_Training/**/**/*_seg.nii')
seg_2018 = []
for img in seg_path:
    image_seg=nib.load(img).get_fdata()
    seg_2018.append(image_seg)
len(seg_2018)
```

```python
#load all t1-contrast images from BRATS 2019
t1_path = glob.glob('MICCAI_BraTS_2019_Data_Training/**/**/*_t1.nii')
t1_2019 = []
for img in t1_path:
    image_t1=nib.load(img).get_fdata()
    t1_2019.append(image_t1)
len(t1_2019)
```

```python
#load all flair contrast images from BRATS 2019
flair_path = glob.glob('MICCAI_BraTS_2019_Data_Training/**/**/*_flair.nii')
flair_2019 = []
for img in flair_path:
    image_flair=nib.load(img).get_fdata()
    flair_2019.append(image_flair)
len(flair_2019)
```

```python
# load segmented ground truth from BRATS 2019
seg_path = glob.glob('MICCAI_BraTS_2019_Data_Training/**/**/*_seg.nii')
seg_2019 = []
for img in seg_path:
    image_seg=nib.load(img).get_fdata()
    seg_2019.append(image_seg)
len(seg_2019)
```

```python
# plot flair image and mask (BRATS 2019)
seg_volume = seg_2019[0]
flair_volume = flair_2019[0]
mid_idx = seg_volume.shape[2] // 2

# Access the middle slice from the selected volume
flair_mid_slice = flair_volume[:,:,mid_idx]
seg_mid_slice = seg_volume[:,:,mid_idx]

fig, ax = plt.subplots(1, 2, figsize=(12, 6)) 

ax[0].imshow(flair_mid_slice, cmap='gray')
ax[0].set_title('FLAIR MRI Slice')
ax[0].axis('off')

ax[1].imshow(seg_mid_slice, cmap='gray')
ax[1].set_title('Segmented tumor')
ax[1].axis('off') 

plt.show()
```
